```python
import pandas as pd
```

```python
# TODO: fare il merge per bene
TRAIN_PATH = './dataset/WASSA22_essay_level_train.tsv'
DEV_PATH = './dataset/WASSA22_essay_level_dev.tsv'

TRAIN_HOPE_PATH = './hope_dataset/english_hope_dev.csv'
DEV_HOPE_PATH = './hope_dataset/english_hope_dev.csv'

dev_df = pd.read_csv(TRAIN_PATH, sep='\t')
dev_df = pd.read_csv(DEV_PATH, sep='\t')

train_hope_df = pd.read_csv(TRAIN_HOPE_PATH, sep='\t')
dev_hope_df = pd.read_csv(DEV_HOPE_PATH, sep='\t')
```

```python
dev_hope_df
```

```python
columns = train_df.columns.tolist()
columns
```

```python
def new_df(old_df, columns=columns):
    # Creazione del nuovo DataFrame
    new_df = pd.DataFrame(columns=columns)

    # Copia dei valori 'essay' e 'emotion' dal DataFrame originale
    new_df['essay'] = old_df['essay']
    new_df['emotion'] = old_df['emotion']

    # Impostazione dei valori 'missing' per le altre colonne
    other_columns = [col for col in columns if col not in ['essay', 'emotion']]
    new_df[other_columns] = 'missing'

    return new_df
```

```python
new_train_hope_df = new_df(train_hope_df)
new_dev_hope_df = new_df(dev_hope_df)
```

```python
new_dev_hope_df
```

```python
new_train_df = pd.concat([train_df, new_train_hope_df])

new_train_df['essay'] = new_train_df['essay'].replace("’", "'")
new_train_df['essay'] = new_train_df['essay'].replace("‘", "'")

new_train_df = new_train_df.sample(frac=1).reset_index(drop=True)

new_train_df
```

```python
new_dev_df = pd.concat([dev_df, new_dev_hope_df])

new_dev_df['essay'] = new_dev_df['essay'].replace("’", "'")
new_dev_df['essay'] = new_dev_df['essay'].replace("‘", "'")

new_dev_df = new_dev_df.sample(frac=1).reset_index(drop=True)
```

```python
TRAIN_SAVE_PATH = './hope_dataset/WASSA22_essay_level_train.tsv'
DEV_SAVE_PATH = './hope_dataset/WASSA22_essay_level_dev.tsv'

new_train_df.to_csv(TRAIN_SAVE_PATH, index=False, sep='\t')
new_dev_df.to_csv(DEV_SAVE_PATH, index=False, sep='\t')
```
