```python
import pandas as pd
```

```python
emotions = [
    'fear',
    'anger',
    'surprise',
    'sadness',
    'disgust',
    'joy',
    'hope',
    'neutral'
]
```

```python
preds = {}
for emotion in emotions:
	preds[emotion] = pd.read_csv(f'./dataset_binary/logits/{emotion}.tsv', sep='\t', header=None)
n_samples = len(preds['sadness'])
```

```python
preds_list = []
for i in range(n_samples):
    emos = []
    for emotion in emotions:
        if preds[emotion][0].loc[i]==emotion:
            emos.append(emotion)
    preds_list.append(emos)
print(preds_list)

preds_str=[]
for pred in preds_list:
    str=''
    for i, emo in enumerate(pred):
        if i==len(pred)-1:
            str+=emo
        else:
            str+=emo+'/'
    preds_str.append(str)
preds_str
```

```python
dev = pd.read_csv('./datasets/WASSA23_essay_level_dev_preproc.tsv', sep='\t')
```

```python
from utils import compute_metrics
scores = compute_metrics(dev.emotion.tolist(), preds_str, "EMO")
scores
```
