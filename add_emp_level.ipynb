```python
import pandas as pd
from utils import get_emp_levels, generate_prompt, get_emp_levels
```

```python
emp_predictions_path = './empathy_data/predictions_EMP.tsv'
test_df = pd.read_csv('./empathy_data/WASSA23_essay_level_test.tsv', sep='\t')
```

```python
emp_predictions = pd.read_csv(emp_predictions_path, names=['empathy', 'distress'], sep='\t')
emp_predictions
```

```python
test_df = get_emp_levels(test_df, emp_predictions['empathy'], emp_predictions['distress'])
```

```python
test_df
```

```python
for idx, row in test_df.iterrows():
    _, text_prompt_emp, _ = generate_prompt(
                                row['essay'],
                                row['article_id'],
                                row['gender'],
                                row['education'],
                                row['race'],
                                row['age'],
                                row['income'],
                                row['empathy_level'],
                                row['distress_level']
                                )
    essay = row['essay']
    test_df.at[idx, "essay"] = essay + " " + text_prompt_emp
```

```python
test_df.to_csv('./empathy_data/WASSA23_essay_level_test_emp.tsv', sep='\t', index=False)
```
