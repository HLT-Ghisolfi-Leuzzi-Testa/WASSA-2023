```python
import pandas as pd
```

```python
# read training essay level data
hope_data = pd.read_csv('./datasets/english_hope_train.csv', header=None, names=['essay', 'hope', '-'], sep="\t")
hope_data = hope_data.drop(['-'], axis=1)
hope_data
```

```python
hope_data.loc[69].essay # emoji: √∞\
```

```python
for idx, row in hope_data.iterrows():
    if row['hope'] == 'Non_hope_speech':
        hope_data = hope_data.drop([idx])

hope_data = hope_data.drop(['hope'], axis=1)
hope_data
```

```python
import emoji

def emoji_count(s):
    return emoji.emoji_count(s)

for idx, row in hope_data.iterrows():
    if (emoji_count(row['essay']) > 0) or '@' in row['essay']:
        hope_data = hope_data.drop([idx])

hope_data
```

```python
hope_data['essay_len'] = hope_data['essay'].apply(lambda x: len(x.split())) # words count for each essay
hope_data
```

```python
# parole max = 160, min = 40

for idx, row in hope_data.iterrows():
    if row['essay_len'] < 40 or row['essay_len'] > 160:
        hope_data = hope_data.drop([idx])
hope_data 
```

```python
hope_data['essay'] = hope_data['essay'].replace("‚Äô", "'")
hope_data['essay'] = hope_data['essay'].replace("‚Äò", "'")
hope_data
```
