```python
import pandas as pd

TRAIN_22_DATA_PATH = './datasets/WASSA22_train.tsv'
DEV_22_DATA_PATH = './datasets/WASSA22_dev.tsv'
DEV_22_GOLD_PATH = './datasets/goldstandard_dev_2022.tsv'

train_22_df = pd.read_csv(TRAIN_22_DATA_PATH, sep='\t')

dev_22_df = pd.read_csv(DEV_22_DATA_PATH, sep='\t')
dev_22_gold = pd.read_csv(DEV_22_GOLD_PATH, header=None, usecols=[2], names=['emotion'], sep='\t')
dev_22_df = dev_22_df.merge(dev_22_gold, left_index=True, right_index=True, how='outer')

essay_22_df = pd.concat([train_22_df[['essay', 'emotion']], dev_22_df[['essay', 'emotion']]])


DATA2023_PATH = "./datasets/train_essay_level_preproc.tsv"
train2023_df = pd.read_csv(DATA2023_PATH, sep='\t')
```

```python
HOPE_PATH = './datasetS/english_hope.csv'
hope_df = pd.read_csv(HOPE_PATH, sep="\t")

essay_22_df = pd.concat([essay_22_df [['essay', 'emotion']], hope_df[['essay', 'emotion']]])
essay_22_df
```

```python
essay_22_df['essay'] = essay_22_df['essay'].replace("’", "'")
essay_22_df['essay'] = essay_22_df['essay'].replace("‘", "'")
```

```python
essay_22_df = essay_22_df.sample(frac=1).reset_index(drop=True)
essay_22_df
```

```python
DATA_PATH = './datasets/train_essay_level_additional.csv'
essay_22_df.to_csv(DATA_PATH, index=False, sep='\t')
```
